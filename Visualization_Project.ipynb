Data Visualization

```python
import matplotlib . pyplot  as plt 
```

```python

region_sales = d1.groupby('Region')['Sales'].sum().sort_values(ascending=False)
plt.bar(region_sales.index, region_sales.values, color='skyblue',edgecolor='black')

plt.title('Sales Performance by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')


for i, value in enumerate(region_sales.values):
  plt.text(i, value + 1000, f'{int(value)}')

plt.show()
```

```python
region_profit = d1.groupby('Region')['Profit'].sum().sort_values()


plt.bar(region_profit.index, region_profit.values, color='lightgreen', edgecolor='black')

plt.title('Profit by Region')
plt.xlabel('Region')
plt.ylabel('Total Profit')
for i, value in enumerate(region_profit.values):
  plt.text(i, value + 1000, f'{int(value)}')



plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Category', y='Profit', data=d1, estimator=sum)
plt.title("Profit by Product Category")
plt.show()
```

```python
category_sales = d1.groupby('Category')['Sales'].sum()

plt.figure()
plt.pie(category_sales, labels=category_sales.index , autopct='%1.1f%%')
plt.title('Sales Distribution by Category')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Discount', y='Profit', data=d1)
plt.title("Profit vs Discount")
plt.show()
```

```python

sub_sales = d1.groupby('Sub-Category')['Sales'].sum().sort_values()
sns.barplot(x=sub_sales.index, y=sub_sales.values)
plt.title("Sales by Sub-Category")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(x='Category', y='Quantity', data=d1)
plt.title("Quantity Distribution by Category")
plt.show()
```

```python

sns.heatmap(d1[['Sales','Quantity','Discount','Profit']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

Discount is strongly negatively correlated with Profit.

```python
top_customers = d1.groupby('Customer Name')['Profit'].sum().nlargest(10)
sns.barplot(y=top_customers.index, x=top_customers.values)
plt.title("Top 10 Most Profitable Customers")
plt.xlabel("Profit")
plt.show()
```

```python

state_sales = d1.groupby('State')['Sales'].sum().sort_values(ascending=False).head(15)
sns.barplot(x=state_sales.values, y=state_sales.index)
plt.title("Top 15 States by Sales")
plt.xlabel("Sales")
plt.show()
```

Conclusion: The business should enhance operations in strong regions, reduce high discounts, focus on profitable product categories, and prepare stock for seasonal sales to maximize total profit.
